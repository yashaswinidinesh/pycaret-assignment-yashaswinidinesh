```

import pandas as pd
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/00292/Wholesale%20customers%20data.csv"
df = pd.read_csv(url)

from pycaret.clustering import setup, create_model, assign_model, plot_model, save_model

exp = setup(data=df, session_id=42, use_gpu=True, silent=True)
kmeans = create_model("kmeans")
plot_model(kmeans, plot="elbow")
clusters = assign_model(kmeans)
clusters.head()

save_path = save_model(kmeans, "wholesale_kmeans")
print("Saved pipeline at:", save_path)
```
