# Use a public mirror of the well-known creditcard.csv (ULB dataset)
import pandas as pd
url = "https://figshare.com/ndownloader/files/44922021"  # creditcard.csv mirror file on figshare
df = pd.read_csv(url)

# Sample for speed (dataset can be large)
df = df.sample(50000, random_state=42).reset_index(drop=True)

from pycaret.anomaly import setup, create_model, assign_model, plot_model, save_model

exp = setup(data=df.drop(columns=["Class"], errors="ignore"), session_id=42, use_gpu=True, silent=True)
iforest = create_model("iforest")   # Isolation Forest
assign = assign_model(iforest)
plot_model(iforest, plot="tsne")

save_path = save_model(iforest, "creditcard_iforest")
print("Saved pipeline at:", save_path)
