import pandas as pd, requests, io
xlsx_url = "https://archive.ics.uci.edu/ml/machine-learning-databases/00502/online_retail_II.xlsx"
data = pd.read_excel(xlsx_url, sheet_name="Year 2010-2011")
data = data.dropna(subset=["Invoice", "Description"])

from pycaret.arules import setup, create_model, plot_model

exp = setup(
    data = data,
    transaction_id = "Invoice",
    item_id = "Description",
    ignore_items = ["POSTAGE"]
)

rules = create_model(metric="confidence", threshold=0.6, min_support=0.02)
print(rules.head())

plot_model(rules)
